```python
def project(img, h_img, w_img, h_proj, w_proj):

    (row, col, _) = img
    x = row
    y = col
    z = 0
    print("x,y,z = ",x,y,z)
    Q = np.array([x, y, z], dtype=np.float32)
    X = get_2d_to_3d(Q)
    return X

def get_2d_to_3d(point):

    [x, y, z] = point
    k = round(4 / (x**2 + y**2 + 4), 4)
    size = np.sqrt((k * x)**2 + (k * y)**2 + (k-1)**2)
    print("k = ", k)

    return np.array([k * x, k * y, k-1 ])
```

```python
point_2d = []
point_3d = []

for i in range(len(pnts1[0])):
    x,y,z = pnts[0][i], pnts[1][i], pnts[2][i]
    q = x,y,z
    sphere = project(q,20,20,20,20)
    point_2d.append(i)    
    point_3d.append(sphere)

# print('point_2d = ', point_2d)
print('point_3d = ', point_3d)
```
